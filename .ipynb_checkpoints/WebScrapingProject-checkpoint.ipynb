```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import re
import pandas as pd
```

```python
url = "https://www.carandbike.com/used/cars-for-sale?transmission=Automatic&utm_source=uc&utm_medium=desktop&utm_campaign=hb&utm_content=july-monsoon-madness"
# this will bring the entire HTML code and store it in variable
page = requests.get(url)
print(page)
```

```python
#page.content # this is the raw code scraped from the website
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
carListingWrapper = soup.find(class_="listingCardContainer")
carListingWrapper
```

```python
carTitle =  soup.find_all(class_="ucTitle truncate truncate1")
carMakeYear = []
carModel = []
for i in carTitle:
    title = i.get_text().split(" ")
    model=''
    for j in range(1,len(title)):
        if(j==1):
            model = title[j]
        else:
            model = model + " " + title[j]
    carMakeYear.append(title[0])
    carModel.append(model)
print(carMakeYear)
print(carModel)
```

```python
carPrices =  soup.find_all(class_="ucPrice flex gridJcgridAc")
askingPrice=[]
for i in carPrices:
    askingPrice.append(i.get_text())
print(askingPrice)
```

```python
carImages =  soup.find_all(class_="ucImage imgRound imgFull")
try:
    for i in range(len(carImages)):
        carImages[i] = urljoin(url, carImages[i]['src'])
except:
    pass
print(carImages)
```

```python
carSpecifications = soup.find_all(class_="ucInfo flex pt10 pb10")
carSpecs=[]
for i in carSpecifications:
    getSpecs = i.find_all(class_="ucInfoItem relative")
    specs=''
    for j in range(len(getSpecs)):
        if(j==0):
            specs=getSpecs[j].get_text()
        else:
            specs = specs + " | " + getSpecs[j].get_text()
    carSpecs.append(specs)
print(carSpecs)
```

```python
carEMIDetails = soup.find_all(class_="emi flex gridAc pointer")
carMonthlyEMI=[]
for i in carEMIDetails:
    try:
        i.span.unwrap()
    except:
        pass
    carMonthlyEMI.append(i.get_text().split("/")[0])
print(carMonthlyEMI)
```

```python
carHealthScore = soup.find_all("div",class_=re.compile("subDiv flex gridAc gridJc "))
carHealthRating = []
carHealthStatus = []
carHealthScore
for i in carHealthScore:
    carHealthRating.append(i.get_text())
    try:
        carHealthStatus.append(i.get("class")[4])
    except:
        carHealthStatus.append("NA")
print(carHealthRating)
print(carHealthStatus)
```

```python
carLoc = soup.find_all(class_="ucFootText truncate truncate1")
carLocation=[]
for i in carLoc:
    carLocation.append(i.get_text())
print(carLocation)
```

```python
df = pd.DataFrame.from_dict({"Car Model":carModel, "Year Of Registration": carMakeYear, "Car Specifications": carSpecs, "Location": carLocation, "Condition Rating": carHealthRating, "Condition Status": carHealthStatus, "Asking Price": askingPrice, "Monthly EMI": carMonthlyEMI, "Car Image": carImages}, orient='index')
df = df.transpose()
df
```

```python
df.to_csv("UsedCars.csv",index=False)
```
